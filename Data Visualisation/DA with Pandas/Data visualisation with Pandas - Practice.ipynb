1) import libraries
2) create a sample of 1000 data points in pandas and numpy
3) create a line plot
4) create a cumulative sum
5) create line plot of cumulative sum

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.randn(1000),index=pd.date_range('2022-01-01',periods=1000),columns=['value'])
```

```python
df.head()
```

```python
#line plot
df.plot(figsize=(15,5))
```

```python
#create cumsum
df['value'] = df['value'].cumsum()
```

```python
#plot cumsum
df.plot(figsize=(15,5))
```
